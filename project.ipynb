```python
"Saperating the data labels"

X = data.drop(columns = 'Outcome')
y=data["Outcome"]
```

```python
scaler=StandardScaler()

standard_scaler=scaler.fit_transform(X)

print(standard_scaler)
```

```python
X=standard_scaler
Y=data["Outcome"]

X_train, X_test, Y_train, Y_test=train_test_split(X,Y,test_size=42, stratify=Y, random_state=2)

print(X.shape)
print(X_train.shape)
print(X_test.shape)
```

```python
classfier=svm.SVC(kernel="linear")

classfier.fit(X_train, Y_train)
```

```python
"Accuracy score of training data"
x_pred=classfier.predict(X_train)
x_accuracy=accuracy_score(x_pred,Y_train)
```

```python
print("Accury score of training data is :",x_accuracy)
```

```python
"accuracy_score of testing data"

x_test_pred=classfier.predict(X_test)
x_test_accuracy=accuracy_score(x_test_pred,Y_test)
```

```python
"Accuray score of testing data"
print("Accuracy score of tesing data is : ",x_test_accuracy)          
```
